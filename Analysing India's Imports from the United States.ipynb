```python
# 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2. Importing the Data
imp_data = pd.read_excel("India's Trade with US.xlsx", sheet_name = 'Sheet2', index_col = 0, header = 0)
imp_data
```

```python
# 3. Setting the Columns
imp_data.set_index('Commodity', drop = 'S.No.', inplace=True)
```

```python
imp_data
```

```python
# 4. Removing spaces in column names
imp_data.columns = imp_data.columns.str.replace(' ', '')

# 5. Making the index names in lower case
imp_data.index = imp_data.index.str.lower()

# 6. Remove only trailing spaces
imp_data.index= [item.rstrip() for item in imp_data.index]
imp_data.index = imp_data.index.str.replace(' ', '_')
imp_data.index
```

```python
# 7. Change the datatype to float
imp_data.dtypes
```

```python
cols = imp_data.select_dtypes(exclude=['float']).columns
imp_data[cols] = imp_data[cols].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
# 8. Fillling the nan values with 0
imp_data.fillna(0, inplace = True)
imp_data.round(2)
imp_data
```

```python
# 9. Create a dataframe without total imports
goods_imp_data = imp_data.iloc[:-1]
```

```python
# 15. Plot a line graph of top 5 commodities imported from U.S. by India
top_imp_data1 = goods_imp_data.nlargest(n=5, columns = goods_imp_data.columns)

fig, ax = plt.subplots(figsize = (15,7))

ax.plot(goods_imp_data.columns,
       top_imp_data1.T);

ax.set(title = "Plotting the Top 5 Commodities that India Imported from the US in the past ten years",
      xlabel = "Years",
      ylabel = "Value of Imports in US($)")

ax.legend(['Natural Products like semi-precious stones, coins etc.', 'Nuclear Reactors and Boilers', 'Aircrafts, spacecrafts and parts', 'Electrical Products', 'Mineral Fuel and Oil etc'], loc = "upper left", bbox_to_anchor=(0,1));
```
