```python
# 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2. Importing the data
exp_data = pd.read_excel("India's Trade with US.xlsx", sheet_name = 'Sheet1', index_col = 1, header = 1)
```

```python
exp_data
```

```python
# 3. Removing spaces in column names
exp_data.columns = exp_data.columns.str.replace(' ', '')

# 4. Making the index names in lower case
exp_data.index = exp_data.index.str.lower()

# 5. Remove only trailing spaces
exp_data.index= [item.rstrip() for item in exp_data.index]
exp_data.index = exp_data.index.str.replace(' ', '_')
exp_data.index
```

```python
# 6. Removing the Column 'S.No.' from the table
exp_data.drop("S.No.", axis = 1, inplace = True)
```

```python
exp_data
```

```python
# 7. Changing the datatype to float
exp_data.dtypes
```

```python
cols = exp_data.select_dtypes(exclude=['float']).columns
exp_data[cols] = exp_data[cols].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
# 8. Filling the nan values with 0
exp_data.fillna(0, inplace = True)
exp_data.round(2)
exp_data
```

```python
# 9. Creating a dataframe without total exports
goods_exp_data = exp_data.iloc[:-1]
```

```python
# 10. Find the top 10 products exported to United States by India for the past ten years
goods_exp_data.nlargest(n= 10, columns = goods_exp_data.columns)
```

```python
# 11. Find the top 10 goods exported to U.S. by India in the year 2014-15, 2017-18, 2020-21 and 2023-24
# 2014-15
goods_exp_data.iloc[:, :1].nlargest(n= 10, columns = ["2014-2015"])
```

```python
# 15. Plotting a lone graph of the top  commodities exported by India to the U.S. 
top_exp_data = goods_exp_data.nlargest(n= 7, columns = goods_exp_data.columns)
top_exp_data
```
